```python
#initialize dataset 
#topdir = "/home/esiii_lxe/HV-test/8-18-21/13-5kV-pmts/"
#topdir = "/home/esiii_lxe/HV-test/8-15-21/10kV-glitch-zoomed/"
topdir = "/home/esiii_lxe/HV-test/8-19-21/15kV-pmts/"
d = Dataset.Dataset(topdir)
#load dataset raw data, parsing the files
file_prefixes = ["pmt", "anode"]
date_of_dataset = "8-18-21"
d.load_raw(file_prefixes, date_of_dataset)
```

```python

d.pair_by_timestamp(20)
ax = d.plot_timestamp_differences()
ax.set_title("20 ms allowed time difference")
plt.show()


d.pair_by_timestamp(100)
ax = d.plot_timestamp_differences()
ax.set_title("100 ms allowed time difference")
plt.show()
```

```python
#pair events and vary the allowed dt parameter, plotting losses and std 
allowed_dts = np.linspace(10, 150, 10)
cut_efficiency = []
stds = []
means = []
for _ in allowed_dts:
    retvals = d.pair_by_timestamp(_)
    cut_efficiency.append(1 - (retvals[0] + retvals[1])/max(retvals[2]))
    dts = d.get_timestamp_differences()
    stds.append(np.std(dts))
    means.append(np.mean(dts))

```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(allowed_dts, stds, 'o-', label="standard deviation", linewidth=2)
ax.plot(allowed_dts, means, 'o-', label="mean", linewidth=2)
ax.set_xlabel("cut on time difference during pairing (ms)")
ax.set_ylabel("ms")
ax.legend()
ax2 = ax.twinx()
ax2.patch.set_alpha(0.0)
ax2.plot(allowed_dts, cut_efficiency, label="cut efficiency", color="firebrick", linewidth=2)
ax2.set_ylabel("cut efficiency")
ax2.legend(loc='lower right')
#plt.savefig("allowed_dt_span.png", bbox_inches='tight')
plt.show()
```

# Script for exporting separated file lists for all datasets

```python
timestamps_df = pd.DataFrame() #dataframe holding dataset name and timestamps
file_prefixes = ["pmt", "anode"]
topdir = "/home/esiii_lxe/HV-test/"
dataset_names = ["8-15-21/10kV-glitch", "8-15-21/10kV-pmts",  "8-15-21/5kV-glitch", "8-15-21/5kV-pmts", "8-16-21/10kV-glitch", "8-18-21/13-5kV-pmts", "8-19-21/15kV-pmts/", "8-19-21/12kV-glitch", "8-19-21/13-5kV-pmts", "8-19-21/15kV-glitch", "8-19-21/17-5kV-glitch", "8-19-21/17-5kV-pmts"]
dataset_dates = [_.split('/')[0] for _ in dataset_names]

for i, ds in enumerate(dataset_names):
    e = pd.Series()
    e["name"] = ds
    d = Dataset.Dataset(topdir+ds)
    d.load_raw(file_prefixes, dataset_dates[i])
    seplists = d.get_separated_filelists()
    for pref in file_prefixes:
        e[pref] = seplists[pref]
    
    timestamps_df = timestamps_df.append(e, ignore_index=True)

print(timestamps_df.head)
```
