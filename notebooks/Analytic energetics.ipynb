```python
import numpy as np 
import matplotlib.pyplot as plt 

plt.style.use("~/evanstyle.mplstyle")
```

```python
C = 20 #pF
cq = 1.6e-19

dQs = np.array(np.linspace(0.5, 50, 100)) #pC
Vs = np.array(np.linspace(0, 18, 100)) #kV
Es = Vs/0.285
Qs = C*Vs*1000 #pC

U = 0.5*(Qs**2)/(C*1e6) #uJ
```

```python
fig, ax = plt.subplots()
ax.plot(Vs, U)
ax2 = ax.twiny()
ax2.plot(Es, U, linewidth=0)
ax2.grid(False)
ax2.set_xlabel("Bulk field between electrodes [kV/cm]", labelpad=10)
ax.set_xlabel("Applied voltage [kV]")
ax.set_ylabel("Total energy stored [uJ]")
ax.grid(color='k', alpha=0.3)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(Vs, Qs/1e3)
ax2 = ax.twiny()
ax2.plot(Es, Qs/1e3, linewidth=0)
ax2.grid(False)
ax2.set_xlabel("Bulk field between electrodes [kV/cm]", labelpad=10)
ax.set_xlabel("Applied voltage [kV]")
ax.set_ylabel("Total charge stored [nC]")
ax.grid(color='k', alpha=0.3)
plt.show()
```

```python
dQs = [1, 5, 15, 25, 50]
fig, ax = plt.subplots()
ax2 = ax.twiny()
for dq in dQs:
    dU = dq*Qs/(C*1e6) #uJ
    ax.plot(Vs, dU*1000, label="{:d} pC".format(dq))
    ax2.plot(Es, dU*1000, linewidth=0)


ax2.grid(False)
ax2.set_xlabel("Bulk field between electrodes [kV/cm]", labelpad=10)
ax.set_xlabel("Applied voltage [kV]")
ax.set_ylabel("Partial energy in small deposition [nJ]")
ax.grid(color='k', alpha=0.3)
#ax.set_yscale('log')
plt.show()
```
