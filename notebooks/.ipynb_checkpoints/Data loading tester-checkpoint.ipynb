```python
#get random waveforms and plot them
nev = 10
somewaves = d.get_random_waveforms(nev)
for i, row in somewaves.iterrows():
    fig, ax = plt.subplots(nrows=2, figsize=(12, 8))
    pmttimes = np.arange(0, len(row['pmt0-data'])*row['pmtSamplingPeriod'], row['pmtSamplingPeriod'])
    anodetimes = np.arange(0, len(row['anode0-data'])*row['anodeSamplingPeriod'], row['anodeSamplingPeriod'])
    ax[0].plot(pmttimes, row['pmt0-data'], linewidth=2, label="ch 1")
    ax[0].plot(pmttimes, row['pmt0-data'], linewidth=2, label="ch 1")
    ax[1].plot(anodetimes, row['anode0-data'], label="glitch")
    ax[0].legend()
    ax[1].legend()
    #ax[0].set_xlim([35, 45])
    ax[0].set_xlabel("us")
    ax[1].set_xlabel("us")
    ax[0].set_ylabel("V")
    ax[1].set_ylabel("V")
    plt.show()
```
